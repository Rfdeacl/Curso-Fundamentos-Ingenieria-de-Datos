```python
"""Importa la libraria numpy"""
import numpy as np
from numpy.ma.core import reshape

"""Genera una matriz con 1 fila y 16 columnas"""
x = np.arange(1,17)
x

"""Redimensiona la matriz a 4 filas y 4 columnas"""
y= x.reshape(4,4)
y
```
